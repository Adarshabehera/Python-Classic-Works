### Carrying out the total Views with the Channel wise Distribution along with the Visualizations :-

```python
dp_1.groupby('Channel')['Views'].sum().sort_values(ascending = False).plot(kind = 'bar', figsize = (10,5))
plt.title("Channel by View Distribution of Ai Courses")
plt.xlabel("Channel_name")
plt.ylabel("Count_of_views")
```

```python
Most_channel_views = pd.DataFrame(dp_1.groupby('Channel')['Views'].sum().sort_values(ascending = False))
def find_channel(channel):
  if channel >= 50000000:
    return "Most_Viewed_Channel"
  if 500000 < channel < 50000000:
    return "Medium_followers_for_the_channel"
  elif channel < 3000000:
    return "Channels_with_less_followers"
```

```python
Most_channel_views['Channel_Categories'] = Most_channel_views.apply(lambda row: find_channel(row['Views']), axis = 1)
Most_channel_views.style.background_gradient(cmap='RdYlGn')
```
